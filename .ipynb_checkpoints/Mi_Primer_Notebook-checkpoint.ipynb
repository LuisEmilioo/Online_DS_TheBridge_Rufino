## 4.Negritas y Cursiva

Para palabras en negritas, la palabra o frase debe estar entre 2 asteriscos al principio y al final: **Palabra o frase**.

Para palabras en cursiva, la palabra o frase debe estar entre un asterisco al principio y al final: *Palabra o frase*.

Para palabras en cursiva y negritas, la palabra o frase debe estar entre 3 asteriscos al principio y al final: ***Palabra o frase***.

Si se deja un espacio entre los asteriscos del principio y la palabra, no se respetara el salto de linea a menos que demos de nuevo enter entre cada linea.
